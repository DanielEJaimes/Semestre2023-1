## Calcular la sensibilidad, especificidad y precisión para k=3 y 5 usando la distancia euclidiana

### Se define una función para calcular la distancia euclidiana

```python
def euclidiana(test,train):
  distancia = 0
  for i in range(len(test)):
      distancia += (test[i] - train[i]) ** 2
  return math.sqrt(distancia)
```

### Se define una función para obtener los k vecinos más cercanos

```python
def prediccion(X_train, y_train, X_test, k):
    # Lista para guardar las predicciones
    y_pred = []
    for i in range(len(X_test)):
      # Lista para guardar todas las distancias calculadas
        distancias = []
        for j in range(len(X_train)):
            distancia = euclidiana(X_test[i], X_train[j]) # Se calcula la distancia euclidiana
            distancias.append((distancia, y_train[j])) # Se agrega la distancia calculada
        distancias.sort() # Se ordena la lista de las distancias
        vecinos = distancias[:k] # Se sacan las k distancias más cortas
        cont = {}
        for vecino in vecinos: # Se realiza la votación para hacer la predicción
            if vecino[1] in cont:
                cont[vecino[1]] += 1
            else:
                cont[vecino[1]] = 1
        y_pred.append(max(cont, key=cont.get)) # Se agrega la predicción a la lista de predicciones
    return y_pred
```

### Se define una función para calcular sensibilidad, especificidad y precision

```python
def matriz(y_test,y_pred):
  true_positives = [0, 0, 0]
  false_negatives = [0, 0, 0]
  true_negatives = [0, 0, 0]
  false_positives = [0, 0, 0]

  for i in range(len(y_test)):
    for j in range(3):
      if y_pred[i]==j and y_test[i]==j:
        true_positives[j] += 1
      elif y_pred[i]!=j and y_test[i]==j:
        false_negatives[j] += 1
      elif y_pred[i]!=j and y_test[i]!=j:
        true_negatives[j] += 1
      elif y_pred[i]==j and y_test[i]!=j:
        false_positives[j] += 1

  confusion = pd.DataFrame({
      "Flor":["Iris-setosa","Iris-versicolor","Iris-virginica"],
      "True positives": true_positives,
      "False negatives":false_negatives,
      "False positives":false_positives,
      "True negatives":true_negatives
  })
  # Cálculo de sensibilidad
  sensitivity_setosa = confusion.iloc[0][1] / (confusion.iloc[0][1] + confusion.iloc[0][2])
  sensitivity_versicolor = confusion.iloc[1][1] / (confusion.iloc[1][1] + confusion.iloc[1][2])
  sensitivity_virginica = confusion.iloc[2][1] / (confusion.iloc[2][1] + confusion.iloc[2][2])
  sensitivity = [sensitivity_setosa,sensitivity_versicolor,sensitivity_virginica]
  confusion["Sensitivity"] = sensitivity
  # Calculo de especificidad
  specificity_setosa = confusion.iloc[0][4] / (confusion.iloc[0][4] + confusion.iloc[0][3])
  specificity_versicolor = confusion.iloc[1][4] / (confusion.iloc[1][4] + confusion.iloc[1][3])
  specificity_virginica = confusion.iloc[2][4] / (confusion.iloc[2][4] + confusion.iloc[2][3])
  specificity = [specificity_setosa,specificity_versicolor,specificity_virginica]
  confusion["Specificity"] = specificity
  #Calculo de precision
  precision_setosa = confusion.iloc[0][1] / (confusion.iloc[0][1] + confusion.iloc[0][3])
  precision_versicolor = confusion.iloc[1][1] / (confusion.iloc[1][1] + confusion.iloc[1][3])
  precision_virginica = confusion.iloc[2][1] / (confusion.iloc[2][1] + confusion.iloc[2][3])
  precision = [precision_setosa,precision_versicolor,precision_virginica]
  confusion["Precision"] = precision
  confusion.fillna(0)
  return confusion
```

### Separación del dataframe para realizar las predicciones

```python
# Se trabajó con una relación 80-20
data_train = flores[:24]
data_test = flores[24:]
# Asignación de los datos para generar el KNN
x_train = data_train.iloc[:,[0,3]].values
y_train = data_train.iloc[:,4].values
# Asignación de los datos para probar el modelo
x_test = data_test.iloc[:,[0,3]].values
y_test = data_test.iloc[:,4].values
```

### K = 3

```python
k=3
y_pred=prediccion(x_train,y_train,x_test,k)
print(y_test)
print(y_pred)
```

### Tabla con TP, FN, FP, TN, sensibilidad, especificidad y precisión para cada flor

```python
tabla = matriz(y_test,y_pred)
tabla
```
